```python
df_salaries = df.filter(regex="^Sal").copy()
df_salaries = pd.concat([df[["Year","Education.Major"]],
                         df_salaries],axis=1)
#df_salaries["Major"] = df["Education.Major"]
df_salaries.rename(columns={"Education.Major":"Major"},inplace=True)
df_salaries.drop("Salaries.Highest",axis=1,inplace=True)
df_salaries.head()
plots = df_salaries[(df_salaries.Major == "Chemistry")
                | (df_salaries.Major =="Psychology")].plot.hist(column="Salaries.Mean",by="Major")
plots[0].get_figure().set_facecolor("white")
plots[1].get_figure().set_facecolor("white")
#df_salaries.groupby("Major").mean().shape
```

```python
df_salaries.head()
temp= df_salaries.query('Major == "Chemistry" or Major == "Chemical Engineering"')
temp = temp.sort_values("Year")[["Year","Major","Salaries.Mean"]]
temp = temp.pivot(index="Year",columns="Major",values="Salaries.Mean")
temp.plot().get_figure().set_facecolor("white")
#temp.set_index("Year",inplace=True)
#temp.plot().get_figure().set_facecolor("white")
```
